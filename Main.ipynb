```python
sys.setrecursionlimit(10 ** 9)
def solve_problem(problem, algorithm, output):
    hg, s0, fs = problem.generate_problem() # Generamos el problema y obtenemos el hipergrafo que representa el problema, el estado inicial, el heurÃ­stico y la polÃ­tica inicial
    p, h = problem.get_initial_policy_and_heuristic(algorithm)
    problem.print_info() # Imprimimos informaciÃ³n del problema
    bpsg_states = None
    t_i = time.time() # Iniciamos contador
    # Ejecutamos algoritmo seleccionado por parÃ¡metros
    if algorithm == 1:
        lao_algorithm = LAO(hg, s0, h, p, problem)
        bpsg_states = lao_algorithm.LAO()
    elif algorithm == 2:
        rlao_algorithm = RLAO(hg, s0, fs, h, p, problem)
        bpsg_states = rlao_algorithm.RLAO()
    elif algorithm == 3:
        ilao_algorithm = ILAO(hg, s0, h, p, problem)
        bpsg_states = ilao_algorithm.ILAO()        
    elif algorithm == 4:
        blao_algorithm = BLAO(hg, s0, fs, h, p, problem)
        bpsg_states = blao_algorithm.BLAO()
    elif algorithm == 5:
        vi_algorithm = Value_Iteration(hg, p, h, gamma = 0.95)
        vi_algorithm.run(hg.states.keys())

    # Finalizamos contador
    t_f = time.time()

    #Imprimimos resultado
    sol = problem.get_solution(p, bpsg_states)
    print("RESULTADO: ")
    print(sol)

    # Imprimimos tiempo usado
    print("Tiempo usado: " + str(t_f - t_i))
    if output == 'S':
        f = open('res.txt', 'w')
        f.write(sol)
        f.close()
```
